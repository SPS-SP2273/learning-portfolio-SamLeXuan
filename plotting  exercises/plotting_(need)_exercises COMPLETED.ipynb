```python
# Ex1
# Some data for plotting
from matplotlib import pyplot as plt
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='maroon', linestyle='dashed',
             marker='o', markerfacecolor='yellow',
             ecolor='black',capsize=5,
             label='Y values with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='lower right')
plt.show()
```

```python
#Ex 2
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 50, 500) 
y = np.sin(x) * np.exp(-x / 10)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of y = sin(x) * e^(-x/10)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values from 0 to 50
x = np.linspace(0, 50, 500)  # 500 points between 0 and 50

# Calculate y values for the function y = sin(x) * e^(-x/10)
y = np.sin(x) * np.exp(-x / 10)

# Plot the function
plt.plot(x, y)
plt.xlabel('x', fontsize=15, color='darkorange')
plt.ylabel('y', fontsize=15, color='darkorange')
plt.title(r'$y = e^{-\frac{x}{10}} \sin(x)$', fontsize=20)
plt.grid(True, alpha=0.25)
plt.ylim(-1, 1)  # Set the y-axis limit
plt.show()
```
